### Task 3 - Visualize One Graph with NetworkX
Pick any one of the four graphs (the most interesting one) and draw it using NetworkX.

```python
# DRAWING WEIGHTED DIRECTED GRAPH

DG = nx.DiGraph()
# Adding edges
for vertice1, edges in wdg.get_adjacency_list().items():
    for vertice2, weight in edges:
        DG.add_edge(vertice1, vertice2, weight=weight)

pos = nx.circular_layout(DG) # get positions
edge_labels = nx.get_edge_attributes(DG, "weight") # get weights

# Drawing graph
plt.figure(figsize=(12, 9))
plt.title("Weighted directed graph")
nx.draw(DG, pos, with_labels=True, font_weight="bold") # draw graph
nx.draw_networkx_edge_labels(DG, pos, edge_labels=edge_labels) # draw weight labels
plt.show()
```
