### Task 1 - Build Four Graph Variants
Using **the same base graph structure as in your written homework (the manual DFS/DFS exercise)**, construct:

- Weighted Directed graph
- Weighted Undirected graph
- Unweighted Undirected graph
- Unweighted Directed graph


```python
wdg = gf.GraphFactory.create_weighted(vertices=10, directed=True)
wug = gf.GraphFactory.create_weighted(vertices=10, directed=False)
udg = gf.GraphFactory.create_unweighted(vertices=10, directed=True)
uug = gf.GraphFactory.create_unweighted(vertices=10, directed=False)
graphs = [wdg, wug, udg, uug]
```

Creating graphs with this structure:  


```python
for graph in graphs:
    graph.add_edge(0, 4)
    graph.add_edge(3, 0)
    graph.add_edge(3, 4)
    graph.add_edge(4, 5)
    graph.add_edge(8, 4)
    graph.add_edge(3, 6)
    graph.add_edge(6, 7)
    graph.add_edge(7, 8)
    graph.add_edge(7, 9)
    graph.add_edge(8, 5)
    graph.add_edge(5, 1)
    graph.add_edge(1, 2)
```
